#### Carregando o DataFrame

```python
import pandas as pd
df = pd.read_csv('spotify_dataset.csv', sep=',')
df.set_index('Index', inplace=True)
df.head()
```

#### Análise e tratamento de dados

```python
df['Artist Followers'] = pd.to_numeric(df['Artist Followers'], errors='coerce')
df['Popularity'] = pd.to_numeric(df['Popularity'], errors='coerce')
df = df.dropna(how='any', axis=0)
df.info()
df.head()
```

## 1. Quais são os 5 Artistas que mais tiveram Números de vezes mapeado? Colunas esperada (Artist, Number of Times Charted)

```python
df_artist = df[["Artist", "Number of Times Charted"]]
df_group_artist = df_artist.groupby("Artist").sum()
table_artist = df_group_artist.sort_values(by=["Number of Times Charted"], ascending=False)
top5_artist = table_artist.head()
display(top5_artist)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = ["Billie Eilish", "Juice WRLD", "Harry Styles", "Ed Sheeran", "Pop Smoke"]
plt.rc('figure', figsize = (15, 5))
plt.bar(x, top5_artist["Number of Times Charted"], color="blue")
```

##### No eixo X o nome do artista, no Y o número de vezes mapeado.

## 3. Quais são os 5 gêneros que contém mais seguidores? Colunas esperada (Genre, Followers)

```python
df_genre = df[["Genre", "Artist Followers"]]
df_genre = df_genre.groupby("Genre").sum()
df_genre = df_genre.sort_values(by="Artist Followers", ascending=False)
df_genre = df_genre["Artist Followers"] / 1000000000
df_genre.head()
```

##### OBS: O resultado está em bilhões para facilitar a compreenção.

##### Várias músicas contém vários gêneros ao mesmo tempo. Isso significa que a pessoa que gosta de uma música que possui vários gêneros provavelmente gosta de todos os gêneros nos quais ela está inscrita. Aqui estão os gêneros agrupados da forma como aparecem no DataFrame, mas acredito que a resposta não seja esta. Acredito que ainda possamos separar cada gênero e contar quantos seguidores cada gênero separado tem.

```python
df_genre = df[["Genre", "Artist Followers"]]
count = 0
generes = {}
chars = "[]'"
for i in df_genre['Genre']:
    for char in chars:
        if char in i:
            i = i.replace(char,"")
    i = i.split(", ")
    for genere in i:
        if genere not in generes.keys():
            generes.update({genere: 0})
            value = df_genre["Artist Followers"].values[count]
            generes[genere] += value
        else:
            value = df_genre["Artist Followers"].values[count]
            generes[genere] += value
    count += 1
    
for i in generes:
    generes[i] /= 1000000000

df_genre = pd.DataFrame(list(generes.items()), columns=["Genre", "Artist Followers"])
df_genre = df_genre.sort_values(by="Artist Followers", ascending=False)
top5_genre = df_genre.head()
```

### Os cinco gêneros com maior seguidores são:

##### OBS: O resultado está em bilhões para facilitar a compreenção.

```python
plt.bar(top5_genre["Genre"], top5_genre["Artist Followers"], color="blue")
```
